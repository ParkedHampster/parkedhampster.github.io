```python
from matplotlib import pyplot as plt

%matplotlib inline
```

```python
# stylizing, this isn't in the examples
import seaborn as sns

sns.set()
```

```python
fig, ax = plt.subplots()
```

```python
shirts_by_people = {
    "Thomas" : 15,
    "Jane" : 3,
    "Carlos" : 7,
    "Jermaine" : 9,
    "Hayle" : 4
}
x = shirts_by_people.keys()
y = shirts_by_people.values()
```

```python
fig, ax = plt.subplots()
ax.bar(x,y)
'';
```

```python
fig, ax = plt.subplots()
ax.bar(x,y)
#==========================#
ax.set_title("Shirts Per Person")
ax.set_ylabel("Number of Shirts")
ax.set_xlabel("Person Name")
'';
```

```python
fig, ax = plt.subplots()
ax.bar(x,y)
#==========================#

ax.set(
    alpha=False,
    title="Shirts Per Person",
    xlabel="Person Name",
    ylabel="Number of Shirts"
)
plt.xticks(rotation=45)
'';
```
